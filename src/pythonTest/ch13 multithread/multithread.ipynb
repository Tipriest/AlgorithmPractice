### 多任务技术

一个程序是一个进程(Progress)，一个进程可以包含多个线程(Thread)，像是迅雷的下载就运用了多线程技术，除了一般的进程和线程，还有一种守护线程，是运行在后台的一种特殊线程，它独立于控制终端并且周期性的执行某种任务，它的作用是为其他线程提供服务，譬如操作系统的垃圾回收站，打印服务等。

多线程模块包括threading,queue模块等

```python
import threading
print("当前活动线程数量:%d"%(threading.active_count()))
print("当前活动线程数量:%d"%(threading.active_count()))
print("当前线程对象:%s"%(threading.current_thread()))
for i in threading.enumerate():
    print(i)
print("主线程对象:%s"%(threading.main_thread()))
print("threading.TIMEOUT_MAX=%d"%(threading.TIMEOUT_MAX))
```

Threading模块提供了Thread, Lock, RLock, Condition, Semaphore, Event等类的支持功能

(1)Thread类

Thread类是threading模块中最常用的功能，通过调用用户指定函数func, 用于独立生成一个活动的线程.调用用户指定函数func(), 可以有两种方法 
1. 是在Thread创建实例对象时，把func以参数形式传递给构造函数;
2. 通过继承Thread类,并重写run()方法，调用func()函数。在Thread的子类中，只允许对__init__()和run()方法进行重写

### 通过抢火车票使用多线程

##### 1. 不使用多线程

```python
from time import *
from datetime import datetime
tickets = [
    ['2018-4-7 8:00',  '北京', '沈阳', 10, 120],
    ['2018-4-7 9:00',  '上海', '宁波',  5, 100],
    ['2018-4-7 12:00', '天津', '北京', 20,  55],
    ['2018-4-7 14:00', '广州', '武汉',  0, 200],
    ['2018-4-7 16:00', '重庆', '西安',  3, 180],
    ['2018-4-7 18:00', '深圳', '上海', 49, 780],
    ['2018-4-7 18:10', '天津', '北京', 20,  55],
]
def buy_ticket(name, nums, data1, start_station):
    i = 0
    sleep(1)
    for get_records in tickets:
        if get_records[0]==data1 and get_records[1]==start_station:
            if get_records[3]>= nums:
                tickets[i][3] = get_records[3] - nums
                print("%s购票成功!"%(name))
                return nums
            else:
                print("%s: No enough tickets available"%(name))
                return -1
        i += 1
    print("%s: No tickets today"%(name))
    return -1
if __name__ == '__main__':
    print("开始时间:", datetime.now())
    result = buy_ticket('张山', 3, '2018-4-7 9:00',  '上海')
    result = buy_ticket('李四', 1, '2018-4-7 14:00', '广州')
    result = buy_ticket('王五', 2, '2018-4-7 9:00',  '上海')
    print("结束时间:", datetime.now())
    print("剩余票数为:")
    for gets in tickets:
        print(gets)
```

#### 使用threading函数方式实现多线程

```python
import threading
from time import *
from datetime import datetime
tickets = [
    ['2018-4-7 8:00',  '北京', '沈阳', 10, 120],
    ['2018-4-7 9:00',  '上海', '宁波',  5, 100],
    ['2018-4-7 12:00', '天津', '北京', 20,  55],
    ['2018-4-7 14:00', '广州', '武汉',  0, 200],
    ['2018-4-7 16:00', '重庆', '西安',  3, 180],
    ['2018-4-7 18:00', '深圳', '上海', 49, 780],
    ['2018-4-7 18:10', '天津', '北京', 20,  55],
]
def buy_ticket(name, nums, data1, start_station):
    i = 0
    sleep(1)
    for get_records in tickets:
        if get_records[0]==data1 and get_records[1]==start_station:
            if get_records[3]>= nums:
                tickets[i][3] = get_records[3] - nums
                print("%s购票成功!"%(name))
                return nums
            else:
                print("%s: No enough tickets available"%(name))
                return -1
        i += 1
    print("%s: No tickets today"%(name))
    return -1
if __name__ == '__main__':
    print("开始时间:", datetime.now())
    t1 = threading.Thread(target = buy_ticket, args=('张山', 3, '2018-4-7 9:00',  '上海'))
    t2 = threading.Thread(target = buy_ticket, args=('李四', 1, '2018-4-7 14:00', '广州'))
    t3 = threading.Thread(target = buy_ticket, args=('王五', 2, '2018-4-7 9:00',  '上海'))
    t1.start()
    t2.start()
    t3.start()
    t1.join()
    t2.join()
    t3.join()
    print("结束时间:", datetime.now())
    print("剩余票数为:")
    for gets in tickets:
        print(gets)
```

##### 采用Threading类方法实现多线程

采用threading函数实现的使用方法是固定的，不够灵活，采用面向对象，使用Threading类的方法可以更灵活

```python
import threading
from time import *
from datetime import datetime
tickets = [
    ['2018-4-7 8:00',  '北京', '沈阳', 10, 120],
    ['2018-4-7 9:00',  '上海', '宁波',  5, 100],
    ['2018-4-7 12:00', '天津', '北京', 20,  55],
    ['2018-4-7 14:00', '广州', '武汉',  0, 200],
    ['2018-4-7 16:00', '重庆', '西安',  3, 180],
    ['2018-4-7 18:00', '深圳', '上海', 49, 780],
    ['2018-4-7 18:10', '天津', '北京', 20,  55],
]
def buy_ticket(name, nums, data1, start_station):
    i = 0
    sleep(1)
    for get_records in tickets:
        if get_records[0]==data1 and get_records[1]==start_station:
            if get_records[3]>= nums:
                tickets[i][3] = get_records[3] - nums
                print("%s购票成功!"%(name))
                return nums
            else:
                print("%s: No enough tickets available"%(name))
                return -1
        i += 1
    print("%s: No tickets today"%(name))
    return -1
class MThread(threading.Thread):
    def __init__(self, target, args):
        super().__init__()
        self.target = target
        self.args = args
    def run(self):
        self.target(*self.args)

if __name__ == '__main__':
    visitor = [
        ('张山', 3, '2018-4-7 9:00',  '上海'),
        ('李四', 1, '2018-4-7 14:00', '广州'),
        ('王五', 2, '2018-4-7 9:00',  '上海')
    ]
    class_do_list = []
    print("开始时间:", datetime.now())
    for get_r in visitor:
        get_one = MThread(buy_ticket, get_r)
        class_do_list.append(get_one)
    for i in range(len(class_do_list)):
        class_do_list[i].start()
    for i in range(len(class_do_list)):
        class_do_list[i].join()
    print("结束时间:", datetime.now())
    print("剩余票数为:")
    for gets in tickets:
        print(gets)
```

#### 线程同步问题

多线程对一个共享数据进行操作，而产生的异常现象叫做条件竞争，这个条件竞争发生在多线程之间不同步的情况下，所谓的线程同步是指采用技术手段，使多线程有序使用共享数据，避免数据出错现象的发生

下面的代码没有加锁也没有产生冒险问题的原因是Cpython中GIL解释器的全局锁的存在

```python
import threading
from time import *
from datetime import datetime
tickets = [
    ['2018-4-7 8:00',  '北京', '沈阳', 10, 120],
    ['2018-4-9 9:00',  '上海', '宁波',990, 100],
    ['2018-4-7 12:00', '天津', '北京', 20,  55],
    ['2018-4-7 14:00', '广州', '武汉',  0, 200],
    ['2018-4-7 16:00', '重庆', '西安',  3, 180],
    ['2018-4-7 18:00', '深圳', '上海', 49, 780],
    ['2018-4-7 18:10', '天津', '北京', 20,  55],
]
def update_ticket(start_station, nums):
    for i in range(len(tickets)):
        tickets[i][3] += nums
def buy_ticket(name, nums, data1, start_station):
    i = 0
    sleep(1)
    for get_records in tickets:
        if get_records[0]==data1 and get_records[1]==start_station:
            if get_records[3]>= nums:
                tickets[i][3] = get_records[3] - nums
                print("%s购票成功!"%(name))
                return nums
            else:
                print("%s: No enough tickets available"%(name))
                return -1
        i += 1
    print("%s: No tickets today"%(name))
    return -1
class MThread(threading.Thread):
    def __init__(self, target, args):
        super().__init__()
        self.target = target
        self.args = args
    def run(self):
        self.target(*self.args)

if __name__ == '__main__':
    class_do_list = []
    print("开始时间:", datetime.now())
    get_one = MThread(target=update_ticket, args=("上海", 5))
    class_do_list.append(get_one)
    for i in range(500):
        get_one = MThread(target=buy_ticket, args=(i, 2, '2018-4-9 9:00',  '上海'))
        class_do_list.append(get_one)
    get_one = MThread(target=update_ticket, args=("上海", 5))
    # class_do_list.append(get_one)
    for i in range(len(class_do_list)):
        class_do_list[i].start()
    for i in range(len(class_do_list)):
        class_do_list[i].join()
    print("结束时间:", datetime.now())
    print("剩余票数为:")
    for gets in tickets:
        print(gets)
```

在使用Lock()锁的过程中，若对一个线程做重复加锁动作，则会导致线程死锁的问题

```python
import threading
def showOK():
    print("OK")
if __name__ == '__main__':
    t1 = threading.Thread(target=showOK)
    lock = threading.Lock()
    lock.acquire()
    lock.acquire()
    t1.start()
    lock.release()
    lock.release()
    print("锁死了吗")
```

要释放死锁的话，只能选择退出解释器

要防止Lock()锁死线程，可以改用RLock()锁,如下所示:

```python
import threading
def showOK():
    print("OK")
if __name__ == '__main__':
    t1 = threading.Thread(target=showOK)
    lock = threading.RLock()
    lock.acquire()
    lock.acquire()
    t1.start()
    lock.release()
    lock.release()
    print("锁死了吗")
```

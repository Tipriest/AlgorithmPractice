```python
import matplotlib.pyplot as plt
import math
import numpy as np
x = np.arange(1.0, 100)
y = np.zeros_like(x)

# 使用enumerate来获取索引和对应的值
for index, value in enumerate(x):
    # 计算y的每个元素，这里假设您想要计算的是y = e^(0.01 * value)
    y[index] = math.exp( - value * 0.3) + np.random.normal(0, 0.01)

# 绘制图像
plt.figure(figsize=(20, 6))
plt.plot(x, y, label='exp(-0.01*i)')
plt.legend()
plt.show()
```

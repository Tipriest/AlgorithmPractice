#### 采用Python爬取网页数据

有了MongoDB数据库, 就可以通过Python+pymongo实现抓取网页数据的存储

```python
import requests
import pymongo
r = requests.get('https://www.taobao.com/')         #获取网页信息
dis = {"_id":2, "context":r.text}                   #把网页以文本方式存入字典变量
from pymongo.mongo_client import MongoClient        #导入访问MongoDB客户端接口功能
mongoClient = MongoClient('localhost', 27017)       #连接本地MongoDB数据库系统
mongoDatabase = mongoClient.site                    #
mongoCollection = mongoDatabase.T_context3
mongoCollection.insert_one(dis)
import pprint
for get in mongoCollection.find({"_id":2}):
    pprint.pprint(get)
```

可以使用正则表达式对网页中的内容进行提取

```python
import requests
r = requests.get('https://www.taobao.com/')         #获取网页信息
import re
pa = re.compile(r'href=\\.+>+(.+?)</a>')
option = pa.findall(r.text)
for get_text in option:
    print(get_text)
```
